```python
from transformers import pipeline
model = pipeline('sentiment-analysis')

model("Hello there , Im rauf and i like to learn ML ")
```

```python
model(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
)
```

```python
model2 = pipeline("zero-shot-classification")
model2(
    "Im coding in the google colab cause ML lib are give some taugh time to my machine ",
    candidate_labels=["coding" , "learning" , "watching"],
)
```

```python
model3 = pipeline('text-generation' , model="distilgpt2")

model3("Hello there im rauf and im a ")
```

```python
model3("hello there im",
    max_length=30,
    num_return_sequences=2,
)
```

```python
model4 = pipeline("fill-mask")
model4("In google colab u can write <mask> ", top_k=2)
```

```python
model5 = pipeline("ner", grouped_entities=True)
model5("My name is Rauf and I start learning at Hugging Face in pakistan.")
```

```python
model5 = pipeline("question-answering")
model5(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask", model="bert-base-uncased")
result = unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])

result = unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])
```

# bias Ex 🙂

```python
biasmodel = pipeline("fill-mask", model="bert-base-uncased")
res = biasmodel("This man works as a [MASK].")
print([r["token_str"] for r in res ])

res = biasmodel("This woman works as a [MASK].")
print([r["token_str"] for r in res])
```
